```python
import cv2
import numpy as np


net = cv2.dnn.readNet("yolov3.weights","yolov3.cfg")

classes = []

with open("coco.txt",'r') as f:
    classes = f.read().splitlines()


# Create a VideoCapture object and read from input file
# If the input is the camera, pass 0 instead of the video file name v1.mp4
cap = cv2.VideoCapture(0)




while (cap.isOpened()):
    ret, img = cap.read()
    height,width,_ = img.shape
    
    blob = cv2.dnn.blobFromImage(img,1/255,(416,416),(0,0,0),swapRB=True,crop =False)
    net.setInput(blob)

    output_layers_names = net.getUnconnectedOutLayersNames()
    layersOutputs = net.forward(output_layers_names)



    boxes = []
    confindences = []
    class_ids = []

    for outputs in layersOutputs:
        for detection in outputs:
            scores = detection[5:]
            class_id = np.argmax(scores)
            confindence = scores[class_id]
            if confindence > 0.5 :
                center_x = int(detection[0]*width)
                center_y = int(detection[1]*height)
                w = int(detection[2]*width)
                h = int(detection[3]*height)
            
                x = int(center_x - w/2)
                y = int(center_y - h/2)
            
                boxes.append([x,y,w,h])
                confindences.append(float(confindence))
                class_ids.append(class_id) 
            
        
    print(len(boxes))
    indexes =  cv2.dnn.NMSBoxes(boxes, confindences,0.5,0.4)
    print(indexes.flatten())
            
    font = cv2.FONT_HERSHEY_PLAIN
    colors = np.random.uniform(0,255,size=(len(boxes),3))

    for i in indexes.flatten():
        x, y, w, h = boxes[i]
        label = str(classes[class_ids[i]])
        confindence = str(round(confindences[i],2))
        color =colors[i]
        cv2.rectangle(img,(x,y),(x+w,y+h),color,2)
        cv2.putText(img,label + " " + confindence, (x,y+20), font, 2,(255,255,255),2)
        
        if ret == True:
            cv2.imshow('Fram',img)
        
            if cv2.waitKey(25) & 0xFF == ord('q'):
                break
        else:
            break
            
cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

# Create a VideoCapture object and read from input file
# If the input is the camera, pass 0 instead of the video file name
cap = cv2.VideoCapture('HillDrive.mp4')

# Check if camera opened successfully
if (cap.isOpened()== False): 
  print("Error opening video stream or file")

# Read until video is completed
while(cap.isOpened()):
  # Capture frame-by-frame
  ret, frame = cap.read()
  if ret == True:

    # Display the resulting frame
    cv2.imshow('Frame',frame)

    # Press Q on keyboard to  exit
    if cv2.waitKey(25) & 0xFF == ord('q'):
      break

  # Break the loop
  else: 
    break

# When everything done, release the video capture object
cap.release()

# Closes all the frames
cv2.destroyAllWindows()
```
